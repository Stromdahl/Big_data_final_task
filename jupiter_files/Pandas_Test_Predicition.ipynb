```python
import pandas as pd
```

```python
train = pd.read_csv('./raw_data/train.csv',nrows=20_000_000)
train.head(20)
```

```python
fmean = train.mean()
train = train.fillna(fmean)
```

```python
features = ['user_id', 'content_id', 'content_type_id',
       'task_container_id', 'user_answer', 'prior_question_had_explanation']
target = ['answered_correctly']
```

```python
X = train[features]
y = train[target]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
predictions_df = pd.DataFrame()
predictions_df['answered_correctly'] = y_test['answered_correctly'].values

classifier = LogisticRegression()
classifier.fit(X_train, y_train.values.ravel())
predictions = classifier.predict(X_test)
print('LogisticRegression:', classifier.score(X_train, y_train))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
predictions_df = pd.DataFrame()
predictions_df['answered_correctly'] = y_test['answered_correctly'].values

classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train.values.ravel())
predictions = classifier.predict(X_test)
print('Decision Tree Result:', classifier.score(X_train, y_train))
```

```python
from sklearn.svm import SVC

predictions_df = pd.DataFrame()
predictions_df['answered_correctly'] = y_test['answered_correctly'].values

classifier = SVC()
classifier.fit(X_train, y_train.values.ravel())
predictions = classifier.predict(X_test)
print('SVC Result:', classifier.score(X_train, y_train))
```
